# Introduction
James Franck and Gustav Hertz ran their mercury gas experiment to demonstrate the excited states of the mercury atom. This experiment proved the discrete quantized states of electrons. Because of this proof the Bohr model was supported. The Franck-Hertz experiment consisted of three separate pieces of equipment. The main unit consisted of a Franck-Hertz tube inside of an oven with a thermometer showing outside the casing. The tube contained the mercury gas to provide the inelastic collisons between the mercury atoms and electrons. 

# Methods and Setup

For this experiment, we used the Franck-Hertz tube, operating unit and oven along with the oscilloscope and thermometer set up. We heated up the the tube to ~190 degrees celcius and set the dials on the operating unit to 40-50 accelerating voltage and 8V for heating voltage as the handout suggested. We messed with the amplitude and reverse bias to get a Franck-Hertz curve to appear on the oscilloscope. We only got 4 peaks and troughs due to the top of the slope "plateauing" off which made any possible points above this area unreadable. We measured the voltage between each consecutive peaks and troughs. So, we got 6 data points total.

For the data between the peaks, we recieved an average of 541.666mV and for the troughs, we recieved a value of 558.333mV, so our average overall value is 550mV. The Franck-Hertz operating unit gives values that are a factor of 10 lower than the standard, so we had to multiply the values by 10, to recieve the right value. So, our energy was 5.5eV.

The frequency and the wavelength can be found with the equations
\begin{equation}
f = E/h
\end{equation}

\begin{equation}
\lambda = c / f
\end{equation}

```python
eV_exp = 5.5
eV_known = 4.9
h = 4.113e-15
c = 3e8

f = eV_exp/h
wl = c/f

percent_error = (eV_exp-eV_known)/eV_known * 100

print(f, "Hz - Frequency")
print(wl, "nm - Wavelength")
print(percent_error, "% - Percent error")
```

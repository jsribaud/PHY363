In the first experiment, the decrease in relative intensity of radiation emmitted from a lightbulb was measured. Figure 1 shows the relationship between the distance and the decrease in the intensity of radiation from the lightbulb. Figure 2 uses the same measurements, but the distance is the inverse square of the measured distance. The graphs differ in shape, with figure 1 appearing like a exponential decrease and figure 2 appears linear. This suggests that the intensity of radiation decreases proportionally to the inverse of the distance change squared.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
data=ascii.read("../../Documents/Physics3/lab2_ks_isl.csv")
dist=data['Distance']
light=data['Rad']
dist2=data['Distance2']
plt.plot(dist, light, 'bs')
plt.title('Figure 1')
plt.xlabel('Distance (cm)')
plt.ylabel('Relative Intensity')
plt.show()

plt.plot(dist2, light, 'bs')
plt.title('Figure 2')
plt.xlabel('Distance (1/cm^2)')
plt.ylabel('Relative Intensity')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
data=ascii.read("../../Documents/Physics3/lab2_ks_isl.csv")
a=data['Area']
d=data['D']
d2=data['D2']
plt.plot(d, a, 'gs')
plt.title('Figure 3')
plt.xlabel('Distance (cm)')
plt.ylabel('Area (cm^2)')
plt.show()

plt.plot(d2, a, 'gs')
plt.title('Figure 4')
plt.xlabel('Distance (1/cm^2)')
plt.ylabel('Area (cm^2)')
plt.show()
```

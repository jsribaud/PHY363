```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from scipy.optimize import curve_fit

data=ascii.read('lab_final_data.txt')

v=data['Vel'] #km/s
d=data['Dist'] #Mpc

ysig=(np.std(v))


def func (x, a, b):
    return a*x + b

popt, pcov=curve_fit(func, d, v, absolute_sigma=ysig)

plt.plot(d, v, 'ro')
plt.plot(d, np.polyval(popt, d), 'b-')
plt.text(11, 11600, r"a = {0:.9f} +/- {1:.9f}".format(popt[0], pcov[0,0]**0.5))
plt.text(0, -2000, 'Figure 1:Hubble law plot.')
plt.ylabel('Velocity (v)')
plt.xlabel('Distance (Mpc)')
plt.show()

print('Hubble Constant=',popt[0],'km/Mpc*s')
```

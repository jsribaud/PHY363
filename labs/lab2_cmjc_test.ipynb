# Abstract

This experiment explores the validity of the Inverse Square Law through the use of light. The intensity of an effect on an object varies by the inverse square of the distance between the object and the source. Particularly, the experiment was performed in two parts. The variations in the intensity of light in relation to its distance from an object was highlighted by implemeting the use of a laser as well as a Stefan-Boltzmann Lamp in each experiment respectively. The Inverse Square Law proved to be valid based on the data collected.

# Introduction

The fact that the intensity of light and other quantities such as sound decreases as the distance (r) gets larger is intutively acceptable. In everyday life, the inverse proportionality of light or sound with distance is experienced frequently. However, the quantitative relationship between the two is interestingly a constant of (1/r^2). The first part of the experiment utilizes a laser as a light source and a chalk board as the object taking effect. The laser was shot through two diffraction gratings oriented at 90 degrees with respect to eachother. This projected a pattern on the chalk board that formed a symmetric square. The area of the square provided a measurement that could be used to determine the intensity of light by comparing the areas at varying distances from the board. The second part of the experiment employed the use of a Stefan-Boltzmann Lamp as a light source and a radiation sensor as the object taking effect. The intensity of the radiation emmitted by the lamp was measured and compared at differing distances. In both parts of the experiment, the measured intensity with relation to distance should reduce to the constant of (1/r^2) and validate the Inverse Square Law.

# Part One: Experiment and Data

For this experiment, the lab set up was composed of a rolling cart with a laser on it. The laser was elevated with level books due to the fact that the chalk board in which we were intending to project the laser on to was too high in relation to the cart. In front of the laser, was a stand with a clamp holding two perpendicular difraction gratings each made up of 13,700 lines. As a test, the laser was shined through the diffraction grating apparatus and was projected onto the chalk board to have nine laser spots that looked perfectly symetric. If the diffraction gratings were not perpendicular, the square made up of nine laser dots would have resembled a parallelogram. Or, if the laser beam was not perpendicular to the diffraction gratings, the laser spots would have been elongated. 

After verifying that the laser-diffraction grating apparatus was set up correctly, the experiment began at ten centimeters distance between the chalk board and the diffraction gratings. The experiment proceededs by increasing the distance between the laser and the diffraction gratings in ten centimeter increments. To measure this distance, a meter stick was held perpendicular from the chalk board to the diffraction gratings. To further ensure that the distance was accurate the meter stick was laid against the level books which acted as the elevation base for the laser. Furthermore, the distance between the board and the diffraction gratings was carefully measured by taking the meter stick measurement from a top view rather than a side view.

The laser spots were projected onto the board from ten centimeters away to one-hundred centimeters away provding ten data points. With each increment, a ruler was used to measure the top right square's base and height. For consistency, the left side vertical part of this square was always used as the height and the bottom of this square was used as a base. The following data was obtained:

```python
from astropy.io import ascii
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=ascii.read("lab2part1.csv")

Distance=data['Distance']
Area=data['Area']


plt.plot(Distance, Area, 'gs')
plt.axis('tight')

plt.grid(True)

plt.title('Area of a Square vs. Distance', fontsize=18)
plt.xlabel('Distance (cm)', fontsize=12)
plt.ylabel('Area of Square (cm^2)', fontsize=12)


z = np.polyfit(Distance, Area, 1)
p = np.poly1d(z)
plt.plot(Distance,p(Distance),"b-")


plt.show()
```

```python
data=ascii.read("lab2part1.csv")

rsqr=data['rsqr']
Area=data['Area']


plt.plot(rsqr, Area, 'gs')
plt.axis('tight')
plt.grid(True)

plt.title('Area of a Square vs. Distance Squared', fontsize=18)
plt.xlabel('Distance Squared (cm^2)', fontsize=12)
plt.ylabel('Area of Square (cm^2)', fontsize=12)

z = np.polyfit(rsqr, Area, 1)
p = np.poly1d(z)
plt.plot(rsqr,p(rsqr),"b-")

plt.show()

```

# Part Two: Experiment and Data

The second experiment was set up by taping a meter stick to a table and lining the filament inside a Stefan-Boltzmann Lamp up to the zero centimeter mark. Also, a radiation sensor was attached to a voltmeter and was lined up level to the lamp. Without turning on the power supply, the radiation sensor was slid against the meter stick in ten centimeter increments beginning with ten centimeters. The voltmeter at each distance from the lamp measured zero mV. This means that there was initally no radiation from outside sources to interfere with the accuarcy of our measurements.

To begin, the power supply was set to ten volts which illuminated the lamp and we measured as we did previously by sliding the senor away from the lamp in ten centimeter increments. Between each set, it is important to record the voltage quickly because the temperature of the sensor could become no longer relatively constant. The sensor was always turned away from the light source between measurements and measurments were taken quickly. The following data was obtained:


```python
data=ascii.read("radiation_sensor.csv")

Rad=data['Rad']
X=data['X']

plt.plot(X, Rad, 'gs')
plt.axis('tight')

plt.grid(True)

plt.title('Radiation Level vs. Distance From Point Source', fontsize=18)
plt.xlabel('Distance From Point Source (cm)', fontsize=12)
plt.ylabel('Radiation Level (mV)', fontsize=12)

z = np.polyfit(X, Rad, 1)
p = np.poly1d(z)
plt.plot(X,p(X),"b-")

plt.show()
```

```python
Invxsqr=data['Invxsqr']

plt.plot(Invxsqr, Rad, 'gs')
plt.axis('tight')

plt.grid(True)

plt.title('Radiation Level vs. Inverse Distance Squared From Point Source', fontsize=18)
plt.xlabel(' Inverse Distance Squared From Point Source (cm^-2)', fontsize=12)
plt.ylabel('Radiation Level (mV)', fontsize=12)

z = np.polyfit(Invxsqr, Rad, 1)
p = np.poly1d(z)
plt.plot(Invxsqr,p(Invxsqr),"b-")

plt.show()
```

# Conclusion

The first experiment yeilded a more linear graph comparing area and distance which implies that our experiment was successful. The discrepencies in our second experiment can be attributed to the fact that the Stephan Boltzmann Lamp was not an actual point source. The intensity of the radiation was more linear as the distance from the sensor was greater. This is consistent with the expected behavior due to the fact that as the distance is increased the lamp acts more like a point source. Based on the data collected during this experiment the inverse square law is determined to be valid.
